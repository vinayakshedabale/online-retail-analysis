NumPy Mathematical Calculator

```python
import numpy as np

print("      Calculator          ")
print("1.Basic Arithmatic Operations ")
print('2.Statistical Operations')
print('3.Array Manipulations')
print('4.Exit')

while True:
    choice = int(input('Enter the choice(1/2/3/4): '))

    if choice == 1:
        print('Enter the arr 1')
        a = [int(i) for i in input().split()]
        print('Enter the arr 2')
        b = [int(i) for i in input().split()]
        a1 = np.array(a)
        b1 = np.array(b)

        print("    Basic  Arithmatic Operations    ")
        print(f'Addition Operations:{a1 + b1}')
        print(f'Subtraction:{a1 - b1}')
        print(f'Multification:{a1 * b1}')
        print(f'Division:{a1 / b1}')
        print(f'Power:{a1 ** b1}')

    elif choice == 2:
        print("       Stastistica Operations       ")
        print('Enter the array')
        arr = [int(j) for j in input().split()]
        arr1 = np.array(arr)

        print(f'Mean:{np.mean(arr1)}')
        print(f'Median:{np.median(arr1)}')
        print(f'Standard Deviation:{np.std(arr1)}')
        print(f'Variance:{np.var(arr1)}')
        print(f'Minimum:{np.min(arr1)}')
        print(f'Maximum:{np.max(arr1)}')
        print(f'Sum:{np.sum(arr1)}')
        print(f'Product:{np.prod(arr1)}')
        print(f'Shape:{np.shape(arr1)}')
        print(f'Size:{np.size(arr1)}')
        print(f'Data Type:{arr1.dtype}')

    elif choice == 3:
        print('     Array Manipulation    ')
        print('1.Reshape')
        print('2.Sort and Reverse')
        print('3.Unique values')
        print('4.To extraction Accessing first and last elements')
        print('5.Filtering values greater than the mean')

        choice1 = int(input('Enter the choice(1/2/3/4/5): '))

        if choice1 == 1:
            print('Enter the array')
            arr = [int(j) for j in input().split()]
            arr1 = np.array(arr)

            print('1.To Convert 1D array to 2D array')
            print('2.To Convert 1D array 3D array')
            print('3.To Convert 2D or 3D to 1D array')
            print('4.Exit')

            choice2 = int(input('Enter the choice(1/2/3/4): '))

            if choice2 == 1:
                print('To Convert 1D array to 2D array')
                r = int(input('Enter the row Number: '))
                c = int(input('Enter the column Number: '))
                print(np.reshape(arr1, (r, c)))

            elif choice2 == 2:
                print('To Convert 1D array 3D array')
                r = int(input('Enter the row Number: '))
                c = int(input('Enter the column Number: '))
                e = int(input('Enter the elements Number: '))
                print(np.reshape(arr1, (r, c, e)))
                print('To Convert 2D or 3D to 1D array')
                print(np.reshape(arr1, (-1)))

            elif choice2 == 3:
                print('To Convert 2D or 3D to 1D array')
                print(np.reshape(arr1, (-1)))
                print(np.flatten(arr1))

            else:
                print('Exit')

        elif choice1 == 2:
            print('Enter the array')
            arr = [int(k) for k in input().split()]
            arr1 = np.array(arr)

            print('To Sort and Reverse')
            print(f'Sorted array:{np.sort(arr1)}')
            print(f'Reversed array:{np.flip(arr1)}')

        elif choice1 == 3:
            print('Unique Values')
            print('Enter the array')
            arr = [int(l) for l in input().split()]
            arr1 = np.array(arr)
            print(np.unique(arr1))

        elif choice1 == 4:
            print('To extraction Accessing first and last elements')
            print('Enter the array')
            arr = [int(m) for m in input().split()]
            arr1 = np.array(arr)
            print(f'first element:{arr1[0]}')
            print(f'last element:{arr1[-1]}')

        elif choice1 == 5:
            print('Filtering values greater than the mean')
            print('Enter the array')
            arr = [int(n) for n in input().split()]
            arr1 = np.array(arr)
            print(arr1[arr1 > np.mean(arr1)])

    elif choice == 4:
        print("Thank you")
        break

    else:
        print("Invalid choice. Please try again.")
```
